```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
sp_data=data.drop(['species'],axis=1)
sp_data.head()
```

```python
species_data=data['species']
species_data.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(sp_data,species_data,test_size=0.33)
```

```python
y_test.head()
```

```python
model=LogisticRegression()
model.fit(x_train,y_train)
model.predict(x_test)
```

```python
accuracy=model.score(x_test,y_test)
print(f"Accuracy:{accuracy:.2f}")
```

```python
modelsvc=SVC()
modelsvc.fit(x_train,y_train)
modelsvc.predict(x_test)
```

```python
accuracy_svc = model.score(x_test, y_test)
print(f"Accuracy: {accuracy_svc:.2f}")
```

```python
modelkn=KNeighborsClassifier()
modelkn.fit(x_train,y_train)
modelkn.predict(x_test)
```

```python
accuracy_knn = modelkn.score(x_train, y_train)
print(f"Accuracy: {accuracy_knn:.2f}")
```

```python
modelgauss = GaussianNB()
modelgauss.fit(x_train, y_train)
modelgauss.predict(x_test)
```

```python
accuracy_gauss = modelgauss.score(x_train, y_train)
print(f"Accuracy: {accuracy_gauss:.2f}")
```

```python
model_decision_tree = DecisionTreeClassifier()
model_decision_tree.fit(x_train, y_train)
model_decision_tree.predict(x_test)
```

```python
acc_decision_tree =model_decision_tree.score(x_train, y_train)
print(f"Accuracy: {acc_decision_tree:.2f}")
```

```python
model_random_forest = RandomForestClassifier(n_estimators=100)
model_random_forest.fit(x_train, y_train)
model_random_forest.predict(x_test)
```

```python
acc_random_forest =model_random_forest.score(x_train, y_train)
print(f"Accuracy: {acc_random_forest:.2f}")
```

```python
models = pd.DataFrame({
    'Model': ['Logistic Regression','SVC','K Neighbors Classifier','Gaussian NB','Decision Tree Classifier',
              'Random Forest Classifier'],
    'Score': [accuracy,accuracy_svc,accuracy_knn,accuracy_gauss,acc_decision_tree,acc_random_forest]})
models.sort_values(by='Score', ascending=False)
```
